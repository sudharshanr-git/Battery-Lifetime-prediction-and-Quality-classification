```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("corr_data_45_pca.csv")
df
```

```python
from ctgan.synthesizers.ctgan import CTGAN

discrete_columns = ['is_baseline_formation','swelling_severity']

model = CTGAN(epochs=500)
model.fit(df, discrete_columns)

synthetic_data = model.sample(60)
synthetic_data
```

```python
more_data= pd.concat([df, synthetic_data], axis=0)
more_data = more_data.sample(frac=1).reset_index(drop=True)
more_data
```

```python
more_data.to_csv("more_corr_data_45_pca.csv", index=False)
```
