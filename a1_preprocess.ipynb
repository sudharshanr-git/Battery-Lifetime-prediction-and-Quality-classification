```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("correlation_data.csv")
df
```

```python
df.drop(columns=['Unnamed: 0.1','Unnamed: 0','channel_number','cellid'],inplace=True)
```

```python
df25 = df[df['is_room_temp_aging'] == 1]  # Rows where col is 1
df25.drop(columns=['is_room_temp_aging'],inplace=True)
df25.reset_index(inplace=True)
df45 = df[df['is_room_temp_aging'] == 0]  # Rows where col is 0
df45.drop(columns=['is_room_temp_aging'],inplace=True)
df45.reset_index(inplace=True)
```

```python
#removing columns with no information
def clean(df):
    print("before removing:",df.shape)
    print("Columns with no information:")
    for i in df.columns:
        if df[i].nunique()==1:
            print(i)
            df.drop(columns=[i],inplace=True)
    print("after removing:",df.shape)
    return df.copy()
```

```python
df25 = clean(df25)
print('\n')
df45 = clean(df45)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

def impute_with_multiple_linear_regression_df(df, col_with_na):
    cols_to_keep = df.columns[df.notna().all()].tolist()
    
    if not cols_to_keep:
        raise ValueError("No complete columns to use for regression.")

    X_train = df.loc[df[col_with_na].notna(), cols_to_keep]
    y_train = df.loc[df[col_with_na].notna(), col_with_na]
    X_missing = df.loc[df[col_with_na].isna(), cols_to_keep]

    model = LinearRegression()
    model.fit(X_train, y_train)

    y_pred = model.predict(X_missing)
    df.loc[df[col_with_na].isna(), col_with_na] = y_pred

    return df[col_with_na].copy()


def fill_null(df):
    columns_with_null=[]
    for i in df.columns:
        if df[i].isnull().sum() != 0:
            print(i,'-',df[i].isnull().sum())
            if df[i].isnull().sum() > 0.5*len(df):
                df.drop(columns=[i],inplace=True)
            else:
                columns_with_null.append(i)
    
    for i in columns_with_null:
        df[i]=impute_with_multiple_linear_regression_df(df, i)
    return df.copy()
```

```python
df25 = fill_null(df25)
print('\n')
df45 = fill_null(df45)
```

```python
df25.to_csv('corr_data_25.csv')
df45.to_csv('corr_data_45.csv')
```
